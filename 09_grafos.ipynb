```python
class Vertice:
    def __init__(self,clave):
        # Llave del Nodo
        self.id = clave
        # Diccionario de adyacencia
        self.conectadoA = {}

    def agregarVecino(self,vecino,ponderacion=0):
        # gregar una conexin del objeto vertice a otro.
        self.conectadoA[vecino] = ponderacion

    def __str__(self):
        # Imprime el diccionario de adyacencia
        # del vertice
        return str( self.id ) + ' conectadoA: ' + \
        str([x.id for x in self.conectadoA])

    def obtenerConexiones(self):
        # Devuelve los vertices del diccionario de adyacencia
        return self.conectadoA.keys()

    def obtenerId(self):
        # Regresa el nombre del vertice
        return self.id

    def obtenerPonderacion(self,vecino):
        # Devuelve el peso de la arista que conecta al vertice 
        # con el argumento (vecino)
        if vecino in self.conectadoA:
            return self.conectadoA[vecino]
        else:
            return None
```

```python
class Grafo:
    def __init__(self):
        # Lista maestra de vertices
        self.listaVertices = {}
        # Numero de vertices
        self.numVertices = 0

    def agregarVertice(self,clave):
        # Agrega un vertice con key=clave
        self.numVertices = self.numVertices + 1
        nuevoVertice = Vertice(clave)
        self.listaVertices[clave] = nuevoVertice
        return nuevoVertice

    def obtenerVertice(self,n):
        if n in self.listaVertices:
            return self.listaVertices[n]
        else:
            return None

    def __contains__(self,n):
        # Regresa True o False dependiendo si 
        # el vertice (n) esta en la lista
        return n in self.listaVertices

    def agregarArista(self,de,a,costo=0):
        # Conecta el vertice (de) con el vertice (a)
        # con un (costo)
        if de not in self.listaVertices:
            nv = self.agregarVertice(de)
        if a not in self.listaVertices:
            nv = self.agregarVertice(a)
        self.listaVertices[de].agregarVecino(self.listaVertices[a], costo)

    def obtenerVertices(self):
        # Regresa los vertices de la lista maestra
        return self.listaVertices.keys()

    def __iter__(self):
        # Regresa un iterador sobre los vertices
        return iter(self.listaVertices.values())
```
